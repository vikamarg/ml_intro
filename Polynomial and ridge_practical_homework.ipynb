### Leave only one feature that has a linear relationship with the target variable.

```python
df1 = df[["Apparent Temperature (C)", "Temperature (C)"]]
```

### Divide the dataset into training and test sets with the ratio 80:20.

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(df1[["Apparent Temperature (C)"]], df1[["Temperature (C)"]], test_size= 0.1, random_state=41)
print(len(df1), len(xtrain ), len(xtest) )
```

### Implement a function fit_ridge_regression

```python
def fit_ridge_regression(data_vector, response_vector, l=1):
    """
    :param data_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is a data point (x)
    :param response_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is label (y)
    :param l: ridge variable
    :return: numpy array beta of coefficients of the regression
    """
    data_vector["ones"] = 1
    X= data_vector
    Y= response_vector
    beta = np.dot(np.linalg.inv(np.dot(X.T, X) + l * np.identity(X.shape[1])) , X.T ).dot(Y)  
    
    return beta
```

### Implement a function fit_polynomial_regression

```python
def fit_polynomial_regression(data_vector, response_vector, degree=2):
    """
    :param data_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is a data point (x)
    :param response_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is label (y)
    :return: numpy array beta of coefficients of the regression
    """
    
    X = np.array([[x**i for i in range(0,degree+1)] for x in np.array(data_vector)])     
    X = X.reshape(len(data_vector), degree+1)
    Y = np.asarray(response_vector ).reshape( len(response_vector), 1   )
    beta = np.dot( np.dot( np.linalg.inv(np.dot(X.T, X)),   X.T ), Y  )

    return beta
```

### Use the resulting train and test sets with the previous 2 functions to fit ridge regression and polynomial regression lines.

```python
b = fit_polynomial_regression(xtrain, ytrain)
print(b)
y_pred_polinom =[sum(b[i] * x ** i for i in range(len(b))) for x in np.array(xtest)]
plt.figure(figsize=(10,7))
plt.scatter(df["Apparent Temperature (C)"], df["Temperature (C)"] , s=0.5)
plt.xlabel("Apparent Temperature (C)")

curve_x = np.sort(np.array(xtest).T).T
y_pred_polinom_curve =[sum(b[i] * x ** i for i in range(len(b))) for x in curve_x]
plt.plot(curve_x, y_pred_polinom_curve, color='red')
plt.show()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df1[["Apparent Temperature (C)"]], df1[["Temperature (C)"]], test_size= 0.1, random_state=41)

b_ridg = fit_ridge_regression(xtrain, ytrain, l=10 )
print(b_ridg)

xtest["ones"] = 1
y_pred_ridg  = np.dot(xtest, b_ridg)

xtrain = xtrain.drop(["ones"], axis=1)
xtest = xtest.drop(["ones"], axis=1)
```

### Calculate RMSE (root mean squared error) for ridge and polynomial regressions

```python
from sklearn.metrics import mean_squared_error
rmse1 = np.sqrt(mean_squared_error(ytest, y_pred_polinom))
print( "RMSE Polinomial>> ",  rmse1 )

rmse2 = np.sqrt(mean_squared_error(ytest, y_pred_ridg))
print( "RMSE Ridge>> ",  rmse2 )
```
